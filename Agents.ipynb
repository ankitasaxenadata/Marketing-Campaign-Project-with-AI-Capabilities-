# Step 4: Build the Agent

```python
# =============================================================
# MARKETING CAMPAIGN AGENT
# =============================================================

agent_code = '''
import os
import pandas as pd
import anthropic
import streamlit as st
from langchain_anthropic import ChatAnthropic
from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain.tools import tool
from langchain_core.prompts import ChatPromptTemplate

# -------------------------------------------------------------
# LOAD DATA
# -------------------------------------------------------------
@st.cache_data
def load_data():
    df = pd.read_csv("marketing_data.csv")
    return df

df = load_data()

# -------------------------------------------------------------
# TOOL 1: Campaign Strategy Recommender
# -------------------------------------------------------------
@tool
def recommend_campaign_strategy(query: str) -> str:
    """Recommends the best campaign strategy based on historical data.
    Use this when user asks about which campaign type, channel, or
    audience to target for best results."""

    summary = {
        "best_campaign_type": df.groupby("Campaign_Type")["ROI"].mean().idxmax(),
        "best_channel": df.groupby("Channel_Used")["ROI"].mean().idxmax(),
        "best_segment": df.groupby("Customer_Segment")["ROI"].mean().idxmax(),
        "best_location": df.groupby("Location")["ROI"].mean().idxmax(),
        "best_duration": df.groupby("Duration_Days")["ROI"].mean().idxmax(),
        "best_gender": df.groupby("Gender")["Conversion_Rate"].mean().idxmax(),
        "best_age_bucket": df.groupby("Age_Bucket")["Conversion_Rate"].mean().idxmax(),
        "avg_roi": round(df["ROI"].mean(), 2),
        "avg_conversion_rate": round(df["Conversion_Rate"].mean(), 4),
    }
    return str(summary)

# -------------------------------------------------------------
# TOOL 2: ROI Predictor
# -------------------------------------------------------------
@tool
def predict_roi(query: str) -> str:
    """Predicts expected ROI for a campaign based on its settings.
    Use this when user asks about expected ROI or performance
    for a specific campaign configuration."""

    roi_by_type = df.groupby("Campaign_Type")["ROI"].mean().to_dict()
    roi_by_channel = df.groupby("Channel_Used")["ROI"].mean().to_dict()
    roi_by_segment = df.groupby("Customer_Segment")["ROI"].mean().to_dict()
    roi_by_duration = df.groupby("Duration_Days")["ROI"].mean().to_dict()

    result = {
        "roi_by_campaign_type": roi_by_type,
        "roi_by_channel": roi_by_channel,
        "roi_by_customer_segment": roi_by_segment,
        "roi_by_duration_days": roi_by_duration,
        "overall_avg_roi": round(df["ROI"].mean(), 2),
        "high_roi_threshold": round(df["ROI"].quantile(0.75), 2),
    }
    return str(result)

# -------------------------------------------------------------
# TOOL 3: Performance Analyzer
# -------------------------------------------------------------
@tool
def analyze_performance(query: str) -> str:
    """Analyzes overall campaign performance across all dimensions.
    Use this when user asks about performance metrics, trends,
    top performers or underperformers."""

    analysis = {
        "total_campaigns": len(df),
        "success_rate": f"{round(df['Campaign_Success'].mean() * 100, 1)}%",
        "avg_ctr": round(df["CTR"].mean(), 4),
        "avg_cost_per_click": round(df["Cost_Per_Click"].mean(), 2),
        "avg_performance_score": round(df["Performance_Score"].mean(), 4),
        "roi_distribution": df["ROI_Bucket"].value_counts().to_dict(),
        "top_performing_segment": df.groupby("Customer_Segment")["Performance_Score"].mean().idxmax(),
        "top_performing_channel": df.groupby("Channel_Used")["Performance_Score"].mean().idxmax(),
        "monthly_roi": df.groupby("Month")["ROI"].mean().to_dict(),
        "best_month": df.groupby("Month")["ROI"].mean().idxmax(),
    }
    return str(analysis)

# -------------------------------------------------------------
# TOOL 4: Budget Allocator
# -------------------------------------------------------------
@tool
def suggest_budget_allocation(query: str) -> str:
    """Suggests how to allocate marketing budget across channels
    and campaign types. Use this when user asks about budget
    distribution or spending recommendations."""

    channel_roi = df.groupby("Channel_Used")["ROI"].mean()
    channel_weights = (channel_roi / channel_roi.sum() * 100).round(1)

    type_roi = df.groupby("Campaign_Type")["ROI"].mean()
    type_weights = (type_roi / type_roi.sum() * 100).round(1)

    segment_cost = df.groupby("Customer_Segment")["Acquisition_Cost"].mean().round(2)
    segment_roi = df.groupby("Customer_Segment")["ROI"].mean().round(2)

    result = {
        "recommended_budget_by_channel_pct": channel_weights.to_dict(),
        "recommended_budget_by_type_pct": type_weights.to_dict(),
        "acquisition_cost_by_segment": segment_cost.to_dict(),
        "roi_by_segment": segment_roi.to_dict(),
        "most_cost_efficient_segment": segment_cost.idxmin(),
        "highest_roi_segment": segment_roi.idxmax(),
    }
    return str(result)

# -------------------------------------------------------------
# AGENT SETUP
# -------------------------------------------------------------
llm = ChatAnthropic(
    model="claude-sonnet-4-20250514",
    anthropic_api_key=os.environ.get("ANTHROPIC_API_KEY"),
    temperature=0
)

tools = [
    recommend_campaign_strategy,
    predict_roi,
    analyze_performance,
    suggest_budget_allocation
]

prompt = ChatPromptTemplate.from_messages([
    ("system", """You are an expert Marketing Campaign Analyst Agent.
    You have access to a dataset of 200,000 marketing campaigns from 2021.

    Your role is to:
    1. Recommend the best campaign strategies based on data
    2. Predict expected ROI for different campaign configurations
    3. Analyze campaign performance across channels, segments and locations
    4. Suggest optimal budget allocation across channels and campaign types

    Always base your answers on the actual data from the tools.
    Be specific, data-driven and actionable in your recommendations.
    Format your responses clearly with key findings and recommendations."""),
    ("human", "{input}"),
    ("placeholder", "{agent_scratchpad}")
])

agent = create_tool_calling_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# -------------------------------------------------------------
# STREAMLIT UI
# -------------------------------------------------------------
st.set_page_config(
    page_title="Marketing Campaign Agent",
    page_icon="📊",
    layout="wide"
)

st.title("📊 Marketing Campaign AI Agent")
st.markdown("Powered by **Claude Anthropic** | Dataset: 200,000 Marketing Campaigns (2021)")
st.divider()

# Sidebar
with st.sidebar:
    st.header("📋 Quick Questions")
    st.markdown("Click any question to get instant insights:")

    questions = [
        "What is the best campaign strategy to maximize ROI?",
        "Which channel should I use for highest conversion rate?",
        "How should I allocate a $50,000 marketing budget?",
        "What is the predicted ROI for an Email campaign on Facebook?",
        "Which customer segment gives the best performance?",
        "Analyze overall campaign performance for 2021",
    ]

    for q in questions:
        if st.button(q, use_container_width=True):
            st.session_state.user_input = q

    st.divider()
    st.markdown("**Dataset Stats:**")
    st.metric("Total Campaigns", "200,000")
    st.metric("Avg ROI", "5.00")
    st.metric("Avg Conversion Rate", "8%")
    st.metric("Success Rate", "49.97%")

# Main Chat Interface
st.subheader("💬 Ask Your Marketing Agent")

if "messages" not in st.session_state:
    st.session_state.messages = []

if "user_input" not in st.session_state:
    st.session_state.user_input = ""

# Display chat history
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

# Chat input
user_input = st.chat_input("Ask anything about your marketing campaigns...")

if user_input or st.session_state.user_input:
    query = user_input or st.session_state.user_input
    st.session_state.user_input = ""

    # Display user message
    with st.chat_message("user"):
        st.markdown(query)
    st.session_state.messages.append({"role": "user", "content": query})

    # Get agent response
    with st.chat_message("assistant"):
        with st.spinner("Analyzing your marketing data..."):
            try:
                response = agent_executor.invoke({"input": query})
                answer = response["output"]
                st.markdown(answer)
                st.session_state.messages.append(
                    {"role": "assistant", "content": answer}
                )
            except Exception as e:
                st.error(f"Error: {str(e)}")
'''

# Save the agent code to a file
with open('app.py', 'w') as f:
    f.write(agent_code)

print("app.py created successfully!")
```

```python
agent_code = '''
import os
import pandas as pd
import streamlit as st
from anthropic import Anthropic

# -------------------------------------------------------------
# LOAD API KEY — Read directly from environment
# -------------------------------------------------------------
ANTHROPIC_API_KEY = os.environ.get("ANTHROPIC_API_KEY", "")

if not ANTHROPIC_API_KEY:
    st.error("❌ ANTHROPIC_API_KEY not found! Please check your setup.")
    st.stop()

# -------------------------------------------------------------
# LOAD DATA
# -------------------------------------------------------------
@st.cache_data
def load_data():
    df = pd.read_csv("/content/drive/MyDrive/marketing_data.csv")
    return df

df = load_data()

# -------------------------------------------------------------
# ANTHROPIC CLIENT
# -------------------------------------------------------------
client = Anthropic(api_key=ANTHROPIC_API_KEY)

# -------------------------------------------------------------
# TOOL FUNCTIONS
# -------------------------------------------------------------
def recommend_campaign_strategy():
    return {
        "best_campaign_type": df.groupby("Campaign_Type")["ROI"].mean().idxmax(),
        "best_channel": df.groupby("Channel_Used")["ROI"].mean().idxmax(),
        "best_segment": df.groupby("Customer_Segment")["ROI"].mean().idxmax(),
        "best_location": df.groupby("Location")["ROI"].mean().idxmax(),
        "best_duration": int(df.groupby("Duration_Days")["ROI"].mean().idxmax()),
        "best_gender": df.groupby("Gender")["Conversion_Rate"].mean().idxmax(),
        "best_age_bucket": df.groupby("Age_Bucket")["Conversion_Rate"].mean().idxmax(),
        "avg_roi": round(df["ROI"].mean(), 2),
        "avg_conversion_rate": round(df["Conversion_Rate"].mean(), 4),
    }

def predict_roi():
    return {
        "roi_by_campaign_type": df.groupby("Campaign_Type")["ROI"].mean().round(3).to_dict(),
        "roi_by_channel": df.groupby("Channel_Used")["ROI"].mean().round(3).to_dict(),
        "roi_by_customer_segment": df.groupby("Customer_Segment")["ROI"].mean().round(3).to_dict(),
        "roi_by_duration_days": df.groupby("Duration_Days")["ROI"].mean().round(3).to_dict(),
        "overall_avg_roi": round(df["ROI"].mean(), 2),
        "high_roi_threshold": round(df["ROI"].quantile(0.75), 2),
    }

def analyze_performance():
    return {
        "total_campaigns": len(df),
        "success_rate": f"{round(df['Campaign_Success'].mean() * 100, 1)}%",
        "avg_ctr": round(df["CTR"].mean(), 4),
        "avg_cost_per_click": round(df["Cost_Per_Click"].mean(), 2),
        "avg_performance_score": round(df["Performance_Score"].mean(), 4),
        "roi_distribution": df["ROI_Bucket"].value_counts().to_dict(),
        "top_performing_segment": df.groupby("Customer_Segment")["Performance_Score"].mean().idxmax(),
        "top_performing_channel": df.groupby("Channel_Used")["Performance_Score"].mean().idxmax(),
        "monthly_roi": df.groupby("Month")["ROI"].mean().round(3).to_dict(),
        "best_month": int(df.groupby("Month")["ROI"].mean().idxmax()),
    }

def suggest_budget_allocation():
    channel_roi = df.groupby("Channel_Used")["ROI"].mean()
    channel_weights = (channel_roi / channel_roi.sum() * 100).round(1)
    type_roi = df.groupby("Campaign_Type")["ROI"].mean()
    type_weights = (type_roi / type_roi.sum() * 100).round(1)
    segment_cost = df.groupby("Customer_Segment")["Acquisition_Cost"].mean().round(2)
    segment_roi = df.groupby("Customer_Segment")["ROI"].mean().round(2)
    return {
        "recommended_budget_by_channel_pct": channel_weights.to_dict(),
        "recommended_budget_by_type_pct": type_weights.to_dict(),
        "acquisition_cost_by_segment": segment_cost.to_dict(),
        "roi_by_segment": segment_roi.to_dict(),
        "most_cost_efficient_segment": segment_cost.idxmin(),
        "highest_roi_segment": segment_roi.idxmax(),
    }

# -------------------------------------------------------------
# AGENT FUNCTION
# -------------------------------------------------------------
def run_agent(user_query):
    context = {
        "strategy": recommend_campaign_strategy(),
        "roi_data": predict_roi(),
        "performance": analyze_performance(),
        "budget": suggest_budget_allocation(),
    }

    system_prompt = f"""You are an expert Marketing Campaign Analyst Agent.
You have access to data from 200,000 marketing campaigns from 2021.

Here is the complete dataset analysis:
{context}

Your role is to:
1. Recommend the best campaign strategies based on data
2. Predict expected ROI for different campaign configurations
3. Analyze campaign performance across channels, segments and locations
4. Suggest optimal budget allocation across channels and campaign types

Always base your answers on the actual data provided above.
Be specific, data-driven and actionable in your recommendations.
Format your responses clearly with key findings and recommendations."""

    response = client.messages.create(
        model="claude-sonnet-4-20250514",
        max_tokens=1500,
        system=system_prompt,
        messages=[{"role": "user", "content": user_query}]
    )

    return response.content[0].text

# -------------------------------------------------------------
# STREAMLIT UI
# -------------------------------------------------------------
st.set_page_config(
    page_title="Marketing Campaign Agent",
    page_icon="📊",
    layout="wide"
)

st.title("📊 Marketing Campaign AI Agent")
st.markdown("Powered by **Claude Anthropic** | Dataset: 200,000 Marketing Campaigns (2021)")
st.divider()

with st.sidebar:
    st.header("📋 Quick Questions")
    st.markdown("Click any question to get instant insights:")

    questions = [
        "What is the best campaign strategy to maximize ROI?",
        "Which channel should I use for highest conversion rate?",
        "How should I allocate a $50,000 marketing budget?",
        "What is the predicted ROI for an Email campaign on Facebook?",
        "Which customer segment gives the best performance?",
        "Analyze overall campaign performance for 2021",
    ]

    for q in questions:
        if st.button(q, use_container_width=True):
            st.session_state.user_input = q

    st.divider()
    st.markdown("**Dataset Stats:**")
    st.metric("Total Campaigns", "200,000")
    st.metric("Avg ROI", "5.00")
    st.metric("Avg Conversion Rate", "8%")
    st.metric("Success Rate", "49.97%")

st.subheader("💬 Ask Your Marketing Agent")

if "messages" not in st.session_state:
    st.session_state.messages = []

if "user_input" not in st.session_state:
    st.session_state.user_input = ""

for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

user_input = st.chat_input("Ask anything about your marketing campaigns...")

if user_input or st.session_state.user_input:
    query = user_input or st.session_state.user_input
    st.session_state.user_input = ""

    with st.chat_message("user"):
        st.markdown(query)
    st.session_state.messages.append({"role": "user", "content": query})

    with st.chat_message("assistant"):
        with st.spinner("Analyzing your marketing data..."):
            try:
                answer = run_agent(query)
                st.markdown(answer)
                st.session_state.messages.append(
                    {"role": "assistant", "content": answer}
                )
            except Exception as e:
                st.error(f"Error: {str(e)}")
'''

with open('/content/app.py', 'w') as f:
    f.write(agent_code)

print("✅ app.py updated!")
```

```python
agent_code = '''
import os
import pandas as pd
import streamlit as st
from groq import Groq

# -------------------------------------------------------------
# LOAD API KEY — Read directly from environment
# -------------------------------------------------------------
GROQ_API_KEY = os.environ.get("GROQ_API_KEY", "")

if not GROQ_API_KEY:
    st.error("❌ GROQ_API_KEY not found! Please check your setup.")
    st.stop()

# -------------------------------------------------------------
# LOAD DATA
# -------------------------------------------------------------
@st.cache_data
def load_data():
    # Ensure the path is correct for your environment
    df = pd.read_csv("/content/drive/MyDrive/marketing_data.csv")
    return df

df = load_data()

# -------------------------------------------------------------
# GROQ CLIENT
# -------------------------------------------------------------
client = Groq(api_key=GROQ_API_KEY)
MODEL_ID = "meta-llama/llama-4-scout-17b-16e-instruct"

# -------------------------------------------------------------
# TOOL FUNCTIONS
# -------------------------------------------------------------
def recommend_campaign_strategy():
    return {
        "best_campaign_type": df.groupby("Campaign_Type")["ROI"].mean().idxmax(),
        "best_channel": df.groupby("Channel_Used")["ROI"].mean().idxmax(),
        "best_segment": df.groupby("Customer_Segment")["ROI"].mean().idxmax(),
        "best_location": df.groupby("Location")["ROI"].mean().idxmax(),
        "best_duration": int(df.groupby("Duration_Days")["ROI"].mean().idxmax()),
        "best_gender": df.groupby("Gender")["Conversion_Rate"].mean().idxmax(),
        "best_age_bucket": df.groupby("Age_Bucket")["Conversion_Rate"].mean().idxmax(),
        "avg_roi": round(df["ROI"].mean(), 2),
        "avg_conversion_rate": round(df["Conversion_Rate"].mean(), 4),
    }

def predict_roi():
    return {
        "roi_by_campaign_type": df.groupby("Campaign_Type")["ROI"].mean().round(3).to_dict(),
        "roi_by_channel": df.groupby("Channel_Used")["ROI"].mean().round(3).to_dict(),
        "roi_by_customer_segment": df.groupby("Customer_Segment")["ROI"].mean().round(3).to_dict(),
        "roi_by_duration_days": df.groupby("Duration_Days")["ROI"].mean().round(3).to_dict(),
        "overall_avg_roi": round(df["ROI"].mean(), 2),
        "high_roi_threshold": round(df["ROI"].quantile(0.75), 2),
    }

def analyze_performance():
    return {
        "total_campaigns": len(df),
        "success_rate": f"{round(df['Campaign_Success'].mean() * 100, 1)}%",
        "avg_ctr": round(df["CTR"].mean(), 4),
        "avg_cost_per_click": round(df["Cost_Per_Click"].mean(), 2),
        "avg_performance_score": round(df["Performance_Score"].mean(), 4),
        "roi_distribution": df["ROI_Bucket"].value_counts().to_dict(),
        "top_performing_segment": df.groupby("Customer_Segment")["Performance_Score"].mean().idxmax(),
        "top_performing_channel": df.groupby("Channel_Used")["Performance_Score"].mean().idxmax(),
        "monthly_roi": df.groupby("Month")["ROI"].mean().round(3).to_dict(),
        "best_month": int(df.groupby("Month")["ROI"].mean().idxmax()),
    }

def suggest_budget_allocation():
    channel_roi = df.groupby("Channel_Used")["ROI"].mean()
    channel_weights = (channel_roi / channel_roi.sum() * 100).round(1)
    type_roi = df.groupby("Campaign_Type")["ROI"].mean()
    type_weights = (type_roi / type_roi.sum() * 100).round(1)
    segment_cost = df.groupby("Customer_Segment")["Acquisition_Cost"].mean().round(2)
    segment_roi = df.groupby("Customer_Segment")["ROI"].mean().round(2)
    return {
        "recommended_budget_by_channel_pct": channel_weights.to_dict(),
        "recommended_budget_by_type_pct": type_weights.to_dict(),
        "acquisition_cost_by_segment": segment_cost.to_dict(),
        "roi_by_segment": segment_roi.to_dict(),
        "most_cost_efficient_segment": segment_cost.idxmin(),
        "highest_roi_segment": segment_roi.idxmax(),
    }

# -------------------------------------------------------------
# AGENT FUNCTION
# -------------------------------------------------------------
def run_agent(user_query):
    context = {
        "strategy": recommend_campaign_strategy(),
        "roi_data": predict_roi(),
        "performance": analyze_performance(),
        "budget": suggest_budget_allocation(),
    }

    system_prompt = f"""You are an expert Marketing Campaign Analyst Agent.
You have access to data from 200,000 marketing campaigns from 2021.

Here is the complete dataset analysis:
{context}

Your role is to:
1. Recommend the best campaign strategies based on data
2. Predict expected ROI for different campaign configurations
3. Analyze campaign performance across channels, segments and locations
4. Suggest optimal budget allocation across channels and campaign types

Always base your answers on the actual data provided above.
Be specific, data-driven and actionable in your recommendations.
Format your responses clearly with key findings and recommendations."""

    # Updated for Groq Chat Completion
    response = client.chat.completions.create(
        model=MODEL_ID,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_query}
        ],
        temperature=0.2, # Lower temperature for better data accuracy
        max_tokens=1500
    )

    return response.choices[0].message.content

# -------------------------------------------------------------
# STREAMLIT UI
# -------------------------------------------------------------
st.set_page_config(
    page_title="Marketing Campaign Agent",
    page_icon="📊",
    layout="wide"
)

st.title("📊 Marketing Campaign AI Agent")
st.markdown(f"Powered by **Groq | {MODEL_ID}**")
st.divider()

with st.sidebar:
    st.header("📋 Quick Questions")
    st.markdown("Click any question to get instant insights:")

    questions = [
        "What is the best campaign strategy to maximize ROI?",
        "Which channel should I use for highest conversion rate?",
        "How should I allocate a $50,000 marketing budget?",
        "What is the predicted ROI for an Email campaign on Facebook?",
        "Which customer segment gives the best performance?",
        "Analyze overall campaign performance for 2021",
    ]

    for q in questions:
        if st.button(q, use_container_width=True):
            st.session_state.user_input = q

    st.divider()
    st.markdown("**Dataset Stats:**")
    st.metric("Total Campaigns", "200,000")
    st.metric("Avg ROI", "5.00")
    st.metric("Avg Conversion Rate", "8%")
    st.metric("Success Rate", "49.97%")

st.subheader("💬 Ask Your Marketing Agent")

if "messages" not in st.session_state:
    st.session_state.messages = []

if "user_input" not in st.session_state:
    st.session_state.user_input = ""

for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

user_input = st.chat_input("Ask anything about your marketing campaigns...")

if user_input or st.session_state.user_input:
    query = user_input or st.session_state.user_input
    st.session_state.user_input = ""

    with st.chat_message("user"):
        st.markdown(query)
    st.session_state.messages.append({"role": "user", "content": query})

    with st.chat_message("assistant"):
        with st.spinner("Analyzing your marketing data..."):
            try:
                answer = run_agent(query)
                st.markdown(answer)
                st.session_state.messages.append(
                    {"role": "assistant", "content": answer}
                )
            except Exception as e:
                st.error(f"Error: {str(e)}")
'''

with open('/content/app.py', 'w') as f:
    f.write(agent_code)

print("✅ app.py updated for Groq Llama 4 Scout!")
```
